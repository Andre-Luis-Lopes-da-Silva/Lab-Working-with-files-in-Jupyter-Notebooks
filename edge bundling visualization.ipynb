```python
import networkx as nx
import holoviews as hv
from holoviews import opts
from holoviews.operation.datashader import bundle_graph, datashade
hv.extension('bokeh')

# Step 1: Create a graph
G = nx.random_geometric_graph(200, radius=0.2)

# Step 2: Compute layout positions (e.g., spring layout)
positions = nx.spring_layout(G)

# Step 3: Convert to Holoviews Graph
hv_graph = hv.Graph.from_networkx(G, positions)

# Step 4: Apply edge bundling
bundled = bundle_graph(hv_graph)

# Step 5: Plot using datashader
plot = datashade(bundled, normalization='log')

# Optional: Customize appearance
plot.opts(
    opts.Overlay(
        width=800, height=800,
        xaxis=None, yaxis=None,
        show_frame=False,
        title="Force-Directed Edge Bundling (Fixed)"
    )
)
```

```python
import networkx as nx
import holoviews as hv
from holoviews.operation.datashader import bundle_graph, datashade
hv.extension('bokeh')

G = nx.les_miserables_graph()
positions = nx.spring_layout(G)

hv_graph = hv.Graph.from_networkx(G, positions)
bundled = bundle_graph(hv_graph)
datashade(bundled).opts(
    width=700, height=700,
    xaxis=None, yaxis=None, show_frame=False,
    title="Edge Bundling - Les Misérables"
)
```
